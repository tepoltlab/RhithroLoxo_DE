```python
import pandas as pd
from ete3 import NCBITaxa
ncbi = NCBITaxa()
```

```python
hits = pd.read_table("../outputs/blast/blast_contam.merged.tsv",header=None)
hits
```

```python
candidate_contam = [2,2157,10239,6157,6231,4751,33630,33090,2763,554915,543769,33634,37909,1458140] #list contam
#bacteria,archaea,viruses,platyhelminthes,nematoda,fungi,alveolata,viridiplantae,rhodophyta,amoebozoa,rhizaria,stramenopiles,rhizocephala,entoniscidae
```

```python
contam = [] #initialize object
for i in hits.index: #loop over indices in df
    contig = hits.loc[i,0]
    taxid = hits.loc[i,13]
    lineage = ncbi.get_lineage(taxid)
    if bool(set(lineage) & set(candidate_contam)):
        contam.append(contig)
        
hits[hits[0].isin(contam)]
with open('contam_list.txt', 'w') as file:
    for i in contam:
        file.write(">"+'%s\n' % i)
```
